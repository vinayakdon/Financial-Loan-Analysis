## Convert date columns to datetime

```python
date_columns = ['issue_date', 'last_credit_pull_date', 'last_payment_date', 'next_payment_date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col])
```

## Feature Engineering: Extract Year and Month for trend analysis

```python
df['issue_month'] = df['issue_date'].dt.month
df['issue_year'] = df['issue_date'].dt.year
df['month_name'] = df['issue_date'].dt.strftime('%b')

print(f"Dataset Loaded: {df.shape[0]} rows, {df.shape[1]} columns")
```

## Total Number of Loan Applications

```python
TotalNumberOfLoanApplications = df['id'].count()
print(f"Total number of loan applications: {TotalNumberOfLoanApplications}")
```

## Month to Date Loan Application for recent month.

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_year'] == latest_year) & (df['issue_month'] == latest_month)]
mtd_loan_application = mtd_data['id'].count()

print(f"MTD Loan Application : {mtd_loan_application}")
```

## Total Funded Amount by Bank to Customer in Millions

```python
TotalFundedAmount = df['loan_amount'].sum()
print(f"Total Funded Amount: ${TotalFundedAmount / 1000000:.2f} Million")
```

## Month to date Total Funded Amount by Bank

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_year'] == latest_year) & (df['issue_month'] == latest_month)]
mtd_total_funded_amount = mtd_data['loan_amount'].sum()

print(f"MTD Total Funded Amount : ${mtd_total_funded_amount / 1000000:.2f} Million")
```

## Total Amount Received by Bank from Customers in Millions

```python
TotalRecoveredAmount = df['total_payment'].sum()
print(f"Total Recovered Amount: ${TotalRecoveredAmount / 1000000:.2f} Milions")
```

## Month to date Total Amount Recovered by Bank

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_year'] == latest_year) & (df['issue_month'] == latest_month)]
mtd_total_recovered_amount = mtd_data['total_payment'].sum()

print(f"MTD Total Recovered Amount : ${mtd_total_recovered_amount / 1000000:.2f} Million")
```

## Average Interest Rate

```python
average_interest_rate = df['int_rate'].mean()*100
print(f"Average Interest Rate: {average_interest_rate:.2f}%")
```

## Average dept to income ratio

```python
average_dti_ratio = df['dti'].mean()*100
print(f"Average dept to income ratio: {average_dti_ratio:.2f}%")
```

# Exproratory Data Analysis

## 1. What is the overall Loan Status distribution? (Portfolio Health)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='loan_status', data=df, palette='viridis')
plt.title('Loan Status Distribution')
plt.show()
```

## 2. What are the top 5 reasons people apply for loans?

```python
plt.figure(figsize=(10, 5))
top_purposes = df['purpose'].value_counts().head(5)
sns.barplot(x=top_purposes.values, y=top_purposes.index, palette='magma')
plt.title('Top 5 Loan Purposes')
plt.show()
```

## 3. Which 10 states have the highest total loan volume? (Geography)

```python
plt.figure(figsize=(12, 5))
state_vol = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=state_vol.index, y=state_vol.values, palette='magma')
plt.title('Top 10 States by Total Loan Amount')
plt.show()
```

## 4. How does Interest Rate correlate with Loan Grade? (Risk Pricing)

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='grade', y='int_rate', data=df, order=sorted(df['grade'].unique()), palette='Spectral')
plt.title('Interest Rate vs. Loan Grade')
plt.show()
```

## 5. Correlation Heatmap - How do metrics relate? (Relationship)

```python
plt.figure(figsize=(10, 8))
numerical_cols = ['annual_income', 'dti', 'installment', 'int_rate', 'loan_amount']
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='inferno', fmt=".2f")
plt.title('Correlation Heatmap of Financial Metrics')
plt.show()
```

## 6. Does Home Ownership impact the likelihood of a loan being "Charged Off"?

```python
plt.figure(figsize=(15, 15))
ct = pd.crosstab(df['home_ownership'], df['loan_status'], normalize='index')
ct.plot(kind='bar', stacked=True, color=['#e74c3c', '#2ecc71', '#3498db'])
plt.title('Loan Status Proportion by Home Ownership')
plt.ylabel('Percentage')
plt.show()
```

## 7. Distribution of Debt-to-Income (DTI) ratio? (Borrower Burden)

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['dti'], bins=30, kde=True, color='purple')
plt.title('Distribution of Debt-to-Income (DTI) Ratio')
plt.show()
```

## 8. Loan Term vs. Interest Rate - Is a longer term more expensive?

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='term', y='int_rate', data=df, palette='Set2')
plt.title('Interest Rate Distribution by Loan Term')
plt.show()
```

## 9. Total Loan Amount by Employment Length - Does job stability equal more credit?

```python
plt.figure(figsize=(12, 5))
emp_order = ['< 1 year', '1 year', '2 years', '3 years', '4 years', '5 years', '6 years', '7 years', '8 years', '9 years', '10+ years']
sns.barplot(x='emp_length', y='loan_amount', data=df, order=emp_order, estimator=sum, ci=None)
plt.title('Total Loan Volume by Employment Duration')
plt.xticks(rotation=45)
plt.show()
```

## 10. Monthly trend by issue date for total funded amount


```python

monthly_funded = ( df.sort_values(by='issue_date').assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y')).groupby('month_name', sort=False)['loan_amount'].sum().div(1_000_1000).reset_index(name='loan_amount_millions') )

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_funded['month_name'], monthly_funded['loan_amount_millions'], color='skyblue', alpha=0.5)
plt.plot(monthly_funded['month_name'], monthly_funded['loan_amount_millions'], color='blue', linewidth=2)


for i, row in monthly_funded.iterrows():
    plt.text(i, row['loan_amount_millions'] + 0.1, f"${row['loan_amount_millions']:.2f}", ha='center', va='bottom', fontsize=9, rotation=0, color='black')


plt.title('Monthly Funded Amount (Millions)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Funded Amount (Millions)', fontsize=12)
plt.xticks(ticks=range(len(monthly_funded)),labels=monthly_funded['month_name'],rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

## Monthly trend by issue date for total amount received

```python
monthly_received = ( df.sort_values(by='issue_date').assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y')).groupby('month_name', sort=False)['total_payment'].sum().div(1000000).reset_index(name='received_amount_millions') )

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_received['month_name'], monthly_received['received_amount_millions'], color='lightgreen', alpha=0.5)
plt.plot(monthly_received['month_name'], monthly_received['received_amount_millions'], color='green', linewidth=2)


for i, row in monthly_received.iterrows():
    plt.text(i, row['received_amount_millions'] + 0.1, f"${row['received_amount_millions']:.2f}", ha='center', va='bottom', fontsize=9, rotation=0, color='black')


plt.title('Monthly Received Amount (Millions)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Received Amount (Millions)', fontsize=12)
plt.xticks(ticks=range(len(monthly_received)),labels=monthly_received['month_name'],rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Monthly Trend - Is the lending market growing?

```python
plt.figure(figsize=(14, 5))
df_2021 = df[df['issue_year'] == 2021]
monthly_trend = df_2021.groupby('month_name')['loan_amount'].count().reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
monthly_trend.plot(kind='line', marker='o', linewidth=3, color='navy')
plt.title('Monthly Loan Application Trend (2021)')
plt.ylabel('Number of Applications')
plt.show()
```
